```python
import os
os.environ["WANDB_DISABLED"] = "true"

from transformers import (
    Seq2SeqTrainingArguments, Seq2SeqTrainer,
    AutoTokenizer, DataCollatorForSeq2Seq,
    MT5ForConditionalGeneration
)
from datasets import load_dataset
import torch
import evaluate
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import re

BATCH_SIZE = 4
LEARNING_RATE = 2e-5
MAX_EPOCHS = 5
prefix = "خلاصه کن: "
device = "cuda" if torch.cuda.is_available() else "cpu"
checkpoint = "google/mt5-base"

def cleanData(text_list):
    cleaned = []
    for text in text_list:
        text = text.strip()
        text = re.sub(r"\s+", " ", text)
        text = re.sub(r"[^\u0600-\u06FFa-zA-Z0-9\s]", "", text)
        cleaned.append(text)
    return cleaned

raw_datasets = load_dataset("pn_summary")
raw_datasets["train"] = raw_datasets["train"].select(range(0, 10000))
raw_datasets["validation"] = raw_datasets["validation"].select(range(0, 400))

article_lens = [len(a.split()) for a in raw_datasets["train"]["article"]]
summary_lens = [len(s.split()) for s in raw_datasets["train"]["summary"]]
MAX_INPUT_LENGTH = int(np.percentile(article_lens, 90))
MAX_TARGET_LENGTH = int(np.percentile(summary_lens, 90))
print(f"MAX_INPUT_LENGTH = {MAX_INPUT_LENGTH} | MAX_TARGET_LENGTH = {MAX_TARGET_LENGTH}")

tokenizer = AutoTokenizer.from_pretrained(checkpoint)
model = MT5ForConditionalGeneration.from_pretrained(checkpoint).to(device)
data_collator = DataCollatorForSeq2Seq(tokenizer, model=model)

def preprocess_function(examples):
    inputs = cleanData([prefix + doc for doc in examples["article"]])
    model_inputs = tokenizer(inputs, max_length=MAX_INPUT_LENGTH, truncation=True, padding="max_length")

    labels = tokenizer(
        cleanData(examples["summary"]),
        max_length=MAX_TARGET_LENGTH,
        truncation=True,
        padding="max_length"
    )
    labels_ids = [
        [(token if token != tokenizer.pad_token_id else -100) for token in label]
        for label in labels["input_ids"]
    ]
    model_inputs["labels"] = labels_ids
    return model_inputs

tokenized_datasets = raw_datasets.map(preprocess_function, batched=True)
tokenized_datasets.set_format(type='torch', columns=['input_ids', 'labels', 'attention_mask'])

rouge = evaluate.load('rouge')
def metric_fn(eval_predictions):
    predictions, labels = eval_predictions
    if isinstance(predictions, tuple):
        predictions = predictions[0]

    predictions = np.where((predictions < 0) | (predictions >= tokenizer.vocab_size), tokenizer.pad_token_id, predictions)
    labels = np.where((labels < 0) | (labels >= tokenizer.vocab_size), tokenizer.pad_token_id, labels)

    decoded_preds = tokenizer.batch_decode(predictions, skip_special_tokens=True)
    decoded_labels = tokenizer.batch_decode(labels, skip_special_tokens=True)

    decoded_preds = [pred.strip() for pred in decoded_preds]
    decoded_labels = [label.strip() for label in decoded_labels]

    scores = rouge.compute(
        predictions=decoded_preds,
        references=decoded_labels,
        rouge_types=["rouge1", "rouge2", "rougeL"],
        use_stemmer=False
    )
    return {
        "Rouge1": scores["rouge1"],
        "Rouge2": scores["rouge2"],
        "RougeL": scores["rougeL"]
    }

training_args = Seq2SeqTrainingArguments(
    output_dir="./checkpoints",
    eval_strategy="epoch",
    learning_rate=LEARNING_RATE,
    per_device_train_batch_size=BATCH_SIZE,
    per_device_eval_batch_size=BATCH_SIZE,
    weight_decay=0.01,
    save_total_limit=3,
    num_train_epochs=MAX_EPOCHS,
    predict_with_generate=True,
    logging_dir="./logs",
    logging_steps=10,
    save_strategy="epoch"
)

trainer = Seq2SeqTrainer(
    model=model,
    args=training_args,
    train_dataset=tokenized_datasets["train"],
    eval_dataset=tokenized_datasets["validation"],
    tokenizer=tokenizer,
    data_collator=data_collator,
    compute_metrics=metric_fn
)

trainer.train()

metrics = trainer.evaluate()
print(" Final results ROUGE:")
print(metrics)

df_log = pd.DataFrame(trainer.state.log_history)
if 'loss' in df_log.columns:
    df_log.plot(x="epoch", y=["loss", "eval_loss"], style="o-")
    plt.title("Training and Validation Loss")
    plt.xlabel("Epoch")
    plt.ylabel("Loss")
    plt.grid()
    plt.show()

for i in range(3):
    sample = raw_datasets["validation"][i]["article"]
    summary = raw_datasets["validation"][i]["summary"]

    input_ids = tokenizer.encode(prefix + sample, return_tensors="pt", truncation=True, max_length=MAX_INPUT_LENGTH).to(device)
    output = model.generate(
        input_ids,
        max_length=100,
        num_beams=4,
        no_repeat_ngram_size=2,
        length_penalty=1.0,
        repetition_penalty=1.2
    )
    generated = tokenizer.decode(output[0], skip_special_tokens=True)
    for j in range(10):
        generated = generated.replace(f"<extra_id_{j}>", "")

    print(f"\n input {i+1}:", sample[:200], "...")
    print("Human summary:", summary)
    print("Summary of the model:", generated)

os.makedirs("./models", exist_ok=True)
torch.save(model.state_dict(), "./models/mt5_finetuned_weights.pth")
tokenizer.save_pretrained("./models")
```

```python
import torch
from transformers import MT5ForConditionalGeneration, AutoTokenizer
from datasets import load_dataset
from rouge_score import rouge_scorer
from tqdm import tqdm

# Specify device
device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')

# Load base model and apply trained weights
model = MT5ForConditionalGeneration.from_pretrained("google/mt5-base")
model.load_state_dict(torch.load("./models/mt5_finetuned_weights.pth"))


model.to(device)
model.eval()
print(" Model loaded successfully.")

# Load tokenizer
tokenizer = AutoTokenizer.from_pretrained('google/mt5-base')

# Load dataset
raw_datasets = load_dataset("pn_summary")

# Extract articles and summaries from the test set (only first 150 for speed)
val_texts = raw_datasets["test"]["article"][:150]
val_summaries = raw_datasets["test"]["summary"][:150]

# Prepare test data
test_data = [{'text': text, 'summary': summary} for text, summary in zip(val_texts, val_summaries)]

# Create ROUGE scorer
scorer = rouge_scorer.RougeScorer(['rouge1', 'rouge2', 'rougeL'], use_stemmer=True, tokenizer=tokenizer)

# Evaluation function
def evaluate(model, tokenizer, data):
    total_scores = {'rouge1': 0, 'rouge2': 0, 'rougeL': 0}

    for example in tqdm(data, desc="🔍 Evaluating"):
        text = example['text']
        summary = example['summary']

        # Preprocess input
        inputs = tokenizer(text, padding='max_length', truncation=True, max_length=512, return_tensors='pt').to(device)
        with torch.no_grad():
            outputs = model.generate(inputs['input_ids'], attention_mask=inputs['attention_mask'],
                                     max_length=128, length_penalty=1.0,  repetition_penalty=1.2)
        generated_summary = tokenizer.decode(outputs[0], skip_special_tokens=True)

        # Calculate ROUGE score
        scores = scorer.score(summary, generated_summary)
        for metric, score in scores.items():
            total_scores[metric] += score.fmeasure

    # Average scores
    num_examples = len(data)
    avg_scores = {metric: score / num_examples for metric, score in total_scores.items()}
    return avg_scores

# Run evaluation
test_scores = evaluate(model, tokenizer, test_data)

# Print results
print(f"\n ROUGE-1: {test_scores['rouge1']:.4f}")
print(f" ROUGE-2: {test_scores['rouge2']:.4f}")
print(f" ROUGE-L: {test_scores['rougeL']:.4f}")


import matplotlib.pyplot as plt

# رسم نمودار ROUGE
metrics = list(test_scores.keys())
scores = [test_scores[m] for m in metrics]

plt.figure(figsize=(8, 5))
bars = plt.bar(metrics, scores, color=['skyblue', 'salmon', 'lightgreen'])
plt.ylim(0, 1)
plt.title("ROUGE scores on test set")
plt.ylabel("Score (F1)")
plt.xlabel("ROUGE Metric")

for bar in bars:
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2.0, yval + 0.01, f"{yval:.4f}", ha='center', va='bottom')

plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()


# Display some sample outputs (qualitative analysis)
print("\n Sample model outputs:")
for i in range(3):
    print(f"\n=== Sample {i+1} ===")
    print(" Article:")
    print(test_data[i]['text'][:300] + "...\n")
    print(" Human summary:")
    print(test_data[i]['summary'])

    inputs = tokenizer(test_data[i]['text'], padding='max_length', truncation=True, max_length=512, return_tensors='pt').to(device)
    with torch.no_grad():
        output = model.generate(inputs['input_ids'], attention_mask=inputs['attention_mask'], max_length=128)
    generated_summary = tokenizer.decode(output[0], skip_special_tokens=True)

    print("Model summary:")
    print(generated_summary)

```
